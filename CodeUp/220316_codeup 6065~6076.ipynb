```python
# 6066
# 3개의 정수(a, b, c)가 입력되었을 때, 짝(even)/홀(odd)을 출력해보자.

a, b, c = input().split()

a = int(a)
b = int(b)
c = int(c)


if a%2==0 :
    print("even")
else :
    print("odd") 

if b%2==0 :
    print("even")
else :
    print("odd") 
    
if c%2==0 :
    print("even")
else :
    print("odd") 
```

```python
# 6067
# 0이 아닌 정수 1개가 입력되었을 때, 음(-)/양(+)과 짝(even)/홀(odd)을 구분해 분류해보자.

a = int(input())

if a < 0:
    if a % 2==0:
        print('A')
    else:
        print('B')
else:
    if a % 2==0:
        print('C')
    else:
        print('D')
```

```python
# 6068
'''
점수(정수, 0 ~ 100)를 입력받아 평가를 출력해보자.

평가 기준
점수 범위 : 평가
 90 ~ 100 : A
 70 ~   89 : B
 40 ~   69 : C
   0 ~   39 : D
로 평가되어야 한다.
'''

a = int(input())

if a > 89:
    print("A")
elif a >69:
    print("B")
elif a >39:
    print("C")
else:
    print("D")
```

```python
# 6069
'''
평가를 문자(A, B, C, D, ...)로 입력받아 내용을 다르게 출력해보자.

평가 내용
평가 : 내용
A : best!!!
B : good!!
C : run!
D : slowly~
나머지 문자들 : what?
'''

a=input()
if a=='A': 
    print("best!!!") 
elif a=='B': 
    print("good!!") 
elif a=='C': 
    print("run!") 
elif a=='D': 
    print("slowly~") 
else: print("what?")

```

```python
# 6070
'''
월이 입력될 때 계절 이름이 출력되도록 해보자.

월 : 계절 이름
12, 1, 2 : winter
  3, 4, 5 : spring
  6, 7, 8 : summer
  9, 10, 11 : fall
'''

a = int(input())

if (a == 12 or a == 1 or a == 2): 
    print("winter") 
elif (a == 3 or a == 4 or a == 5): 
    print("spring") 
elif ( a == 6 or a == 7 or a == 8): 
    print("summer") 
else: print("fall")
```

```python
# 6071
'''
임의의 정수가 줄을 바꿔 계속 입력된다.
-2147483648 ~ +2147483647, 단 개수는 알 수 없다.

0이 아니면 입력된 정수를 출력하고, 0이 입력되면 출력을 중단해보자.
'''

while True :
    a = int(input())
    if a==0 :
        break
    print(a)
```

```python
# 6072
'''
정수(1 ~ 100) 1개가 입력되었을 때 카운트다운을 출력해보자.
'''

a = int(input())

while True :
    print(a)
    a = a-1
    
    if (a==0):
        break
```

```python
# 6073
'''
정수(1 ~ 100) 1개가 입력되었을 때 카운트다운을 출력해보자.
'''

a = int(input())

while True :
    
    a = a-1
    print(a)
    if (a==0):
        break
```

```python
# 6074
'''
영문 소문자(a ~ z) 1개가 입력되었을 때,
a부터 그 문자까지의 알파벳을 순서대로 출력해보자.
'''

c=input()
i = ord('a')
c = ord(c) 

while i<=c: 
    print(chr(i), end=' ') 
    i+=1
```

```python
# 6075
'''
정수(0 ~ 100) 1개를 입력받아 0부터 그 수까지 순서대로 출력해보자.
'''

n=int(input()) 
i=0 

while i<=n: 
    print(i) 
    i+=1
```

```python
# 6076
'''
정수(0 ~ 100) 1개를 입력받아 0부터 그 수까지 순서대로 출력해보자.
'''

n=int(input())

for i in range(n+1):
    print(i)
```
